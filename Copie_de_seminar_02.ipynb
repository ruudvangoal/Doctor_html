# Визуализация данных

Визуализация данных – это
представление данных в
виде, который обеспечивает
наиболее эффективную
работу человека по их
изучению.

Для начала ответим на вопрос: а нужна ли вообще визуализация при анализе данных?

Рассмотрим синтетический пример, созданный в 1973 году английским математиком Фрэнсисом Джоном Энскомбом -- Квартет Энскомба. 

Квартет Энскомба.png

В данном примере представлены 4 выборки: с различными x и y.   
Следует отметить, что ни одна их выборок не повторяется (присутствуют одинаковые значения, но **каждая выборка уникальна**).    
При этом **базовые статистики** для каждой выборок **идентичны** (!):


*   mean x = 9
*   mean y = 11,5
*   выборочная дисперсия x = 11
*   выборочная дисперсия y = 4,125
*   корреляция между x и y = 0,816


Таким образом, если при анализе данных использовать только базовые статистики, то можно прийти к ошибочному выводу о том, что все 4 выборки -- идентичны.   :
Но мы-то с вами знаем, что это не так 😊.

Поэтому давайте обратимся к визуализации данных:

Визуализация.png

После того, как были построены графики, нам придется отказаться от тех выводов, которые мы сделали после расчета базовых статистик.   
В результате анализа графиков можно сделать следующие выводы:


1.   Первый график: присутствует линейная взаимосвязь, некоторый разброс, линия тренда.
2.   Второй график: мы не можем сделать однозначный прогноз о том, как будет себя вести кривая. Можно предположить квадратичную зависимость между переменными.
3.   Третий график: помимо однозначной взаимосвязи двух величин в выборке присутвуют выбросное значение.
4.   Четвертый график: x -- const, y -- меняется, присутсвует выбросное значение.   

Таким образом, данный пример демонстрирует:
*  **важность визуализации** для анализа данных;
*  **влияние выбросов** на статистические показатели.

Возможна ситуация, когда необходимо отобразить два графика на одном поле:

```python
# Создадим переменные
x = data.Age
y1 = data.Total_Bilirubin
y2 = data.Direct_Bilirubin
```

```python
# Построение графика
plt.title("Значения общего и прямого билирубина \
при различном возрасте пациента") # заголовок
plt.xlabel("Возраст пациента")         # ось абсцисс
plt.ylabel("y1, y2")    # ось ординат
plt.grid()              # включение отображение сетки

plt.scatter(x=x, y=y1, color="green") # общий билирубин

plt.scatter(x=x, y=y2, color="blue", marker="x") # прямой билирубин

plt.legend(['Общий билирубин', 'Прямой билирубин']) # легенда

plt.show()
```

Значение общего билирубина всегда больше, чем значение прямого (это и понятно, так как прямой билирубин является одной из составляющих общего билирубина).

Третья, довольно часто встречающаяся задача – это отобразить **два или более различных поля**, на которых будет отображено по одному или более графику.

Построим предыдущий график на разных полях.

```python
# Построение графиков
plt.figure(figsize=(9, 9))

plt.subplot(2, 1, 1)
plt.title("Значения общего и прямого билирубина \
при различном возрасте пациента", fontsize=14) # заголовок
plt.scatter(x, y1)               # построение графика
plt.xlabel("Возраст пациента", fontsize=14)  # ось абсцисс
plt.ylabel("Общий билирубин", fontsize=14) # ось ординат
plt.grid(True)                # включение отображения сетки

plt.subplot(2, 1, 2)
plt.scatter(x, y2, color="green")               # построение графика
plt.xlabel("Возраст пациента", fontsize=14)  # ось абсцисс
plt.ylabel("Прямой билирубин", fontsize=14) # ось ординат
plt.grid(True)                # включение отображения сетки

```

Здесь мы воспользовались новыми функциями: 

**figure()** – функция для задания глобальных параметров отображения графиков. В нее, в качестве аргумента, мы передаем кортеж, определяющий размер общего поля. 

**subplot()** – функция для задания местоположения поля с графиком. Существует несколько способов задания областей для вывода через функцию subplot(), мы воспользовались следующим: первый аргумент – количество строк, второй – столбцов в формируемом поле, третий – индекс (номер поля, считаем сверху вниз, слева направо).

Остальные функции Вам уже знакомы, дополнительно мы использовали параметр **fontsize** для функций xlabel() и ylabel(), для задания размера шрифта.

Мы рассмотрели совсем небольшой пласт функционала matplotlib (подробности можно найти в докуменатции к [ matplotlib.pyplot](https://matplotlib.org)).

#### **График «квантиль-квантиль»** (QQplot, QQ)

График "квантиль-квантиль" используется для проверки распределения выборки.

Этот график генерирует собственную выборку идеализированного распределения, с которым мы сравниваем реальное распределение выбранной переменной. Идеализированные образцы делятся на группы (например, 5), называемые квантилями. Каждая точка данных в выборке связана с аналогичным элементом из идеализированного распределения с тем же кумулятивным распределением.

Результирующие точки построены в виде точечной диаграммы с идеализированным значением на оси x и выборкой данных на оси y.

Идеальное совпадение для распределения будет показано линией точек под углом 45 градусов от нижнего левого угла графика до правого верхнего угла. Часто на графике рисуется линия, чтобы прояснить это ожидание. Отклонения по точкам от линии показывают отклонение от ожидаемого распределения.

Для построения QQ-графика необходимо использовать дополнительные пакеты:

```python
import pylab as py
import scipy.stats as stats
```

```python
stats.probplot(data.Albumin, dist="norm", plot=py)
plt.xlabel("Альбумин") # ось абсцисс
py.grid()
py.show()

stats.probplot(data.Total_Bilirubin, dist="norm", plot=py)
plt.xlabel("Общий билирубин") # ось абсцисс
py.grid()
py.show()
```

Можно утверждать, что переменная Albumin близка к нормальному распределению, а переменная Total_Bilirubin -- нет.

#### **Столбчатая диаграмма**

Диаграмма, представленная прямоугольными зонами, высоты или длины которых пропорциональны величинам, которые они отображают.   
Гистограммы используют для представления **числовых данных одной категории и их изменения с течением времени**.

Перечислим **отличия столбчатых диаграмм от гистограмм**:

1.  Столбчатые диаграммы
*   Есть расстояние между столбцами
*   Используются для сравнения категориальных данных
*   Можно переставлять столбцы местами и смысл не изменится



2.  Гистограммы
*   Нет расстояния между столбцами
*   Если поменять местами, то нарушится последовательная логика

#### **Мозаичная диаграмма**


Графический метод визуализации данных по двум или более качественным переменным. 

Для построения мозаичной диаграммы необходимо импортировать дополнительный пакет:

```python
from statsmodels.graphics.mosaicplot import mosaic
```

Данный функционал работает с dataframe, поэтому создадим отдельный набор данных с категориальными переменными.

```python
data_mosaic = data[["Gender", "Dataset"]]
```

Посторим мозаичную диаграмму

```python
mosaic(data_mosaic, ['Gender', 'Dataset'], gap=0.03)
plt.show()
```

Вывод: мужчины чаще имеют болезни печени, чем женщины.



**Усложним задачу**: построим мозаичную диаграмму, которая будет включать в себя переменные Age и Dataset.    
Для этого необходимо количественную переменную Age сделать категориальной. Наиболее простой способ сделать это -- разделить все возрасты на промежутки:


*   до 30 лет;
*   от 30 до 55 лет;
*   более 50 лет.



С помощью метода loc() добавим в наши исходный dataset столбец age_group:

```python
data.loc[data.Age <= 30, 'age_group'] = 'young'
data.loc[(30 < data.Age) & (data.Age <= 55), 'age_group'] = 'adult'
data.loc[data.Age > 55, 'age_group'] = 'aged'
data.head(10)
```

Посторим мозаичную диаграмму:

```python
mosaic_diag = mosaic(data, ['age_group','Dataset'], gap=0.03)
```

Вывод: Доля больных людей среди молодых меньше, чем среди взрослых и пожилых. Среди взрослых и пожилых людей доля больных практически не отличается.

# Тесты для проверки гипотезы о нормальности распределения 

Рассмотрим следующие **тесты для проверки гипотезы о нормальности распределения**:


1.   Тест Шапиро-Уилка;
2.   Тест χ^2 Пирсона;
3.   Тест Колмогорова-Смирнова в модификации Лиллиефорса;
4.   Тест Андерсона-Дарлинга.




Все перечисленные тесты находятся в пакете scipy.stats.

### Тест Шапиро-Уилка

Тест Шапиро-Уилка оценивает выборку и дает количественную оценку вероятности того, что данные были получены из гауссовского распределения.

```python
from scipy.stats import shapiro
```

В качестве переменной будем рассматривать Albumin.

```python
stat, p = shapiro(data.Albumin)
print('Statistics=%.3f, p=%.3f' % (stat, p))

alpha = 0.05  # уровень значимости

if p > alpha:
	print('Переменная имеет нормальное распределение (не можем отклонить H0))')
else:
	print('Переменная не имеет нормальное распределение (отклоняем H0)')
```

### Тест Колмогорова-Смирнова (в модификации Лиллиефорса)

Критерий однородности Смирнова используется для проверки гипотезы о принадлежности двух независимых выборок одному закону распределения, то есть о том, что два эмпирических распределения соответствуют одному и тому же закону.

Импортируем необходимый модуль:

```python
from scipy.stats import kstest
```

Проводим тест (в качестве закона рассматривем нормальное распределение):

```python
kstest(data.Albumin, 'norm')
```

Поскольку значение p меньше 0,05, мы отвергаем нулевую гипотезу. У нас есть достаточные доказательства того, что выборочные данные не получены из нормального распределения.

### Тест Андрерсона-Дарлинга

Тест представляет собой модифицированную версию более сложного непараметрического статистического критерия Колмогорова-Смирнова.

Особенностью теста Андерсона-Дарлинга является то, что он возвращает список критических значений, а не одно значение p. Это может послужить основой для более тщательной интерпретации результата.

Импортируем необходимый модуль:

```python
from scipy.stats import anderson
```

Рассчитываем тест

```python
result = anderson(data.Albumin)

print('Statistic: %.3f' % result.statistic)

p = 0
for i in range(len(result.critical_values)):
	sl, cv = result.significance_level[i], result.critical_values[i]

	if result.statistic < result.critical_values[i]:
		print('%.3f: %.3f, данные имеют нормальное распределение \
		(не отвергаем H0)' % (sl, cv))
	else:
		print('%.3f: %.3f, данные не имеют нормальное распределение \
		(отвергаем  H0)' % (sl, cv))
```

Таким образом, результаты уже третьего теста дают основания считать, что переменная Albumin распределена не по нормальному закону.

### Тест Д'Агостино

Предназначен для проверки гипотезы о принадлежности анализируемой выборки нормальному закону распределения.


```python
from scipy.stats import normaltest

stat, p = normaltest(data.Albumin)
alpha = 0.05  # уровень значимости

print('p = ', p)

if p > alpha:
	print('Переменная имеет нормальное распредление (не можем отклонить H0))')
else:
	print('Переменная не имеет нормальное распредление (отклоняем H0)')
```
